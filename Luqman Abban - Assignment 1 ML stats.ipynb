5. Calculate the variance and standard deviation for numerical columns.

```python
variance=round(car_sales.select_dtypes(include='number').var()),2
st_deviation=round(car_sales.select_dtypes(include='number').std()),2
print("Variance:\n", variance)
print("\nStandard Deviation:\n", st_deviation)
```

6. For the same columns Calculate IQR and identify outliers in these columns

```python
column_n = ['Sales_in_thousands','Price_in_thousands','Engine_size','Horsepower']
outliers = {}
for column in column_n:
    if column in car_sales.columns:
        Q1 = car_sales[column].quantile(0.25)
        Q3 = car_sales[column].quantile(0.75)

        IQR = Q3 - Q1

        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR

        outlier_indices = car_sales[(car_sales[column] < lower_bound) | (car_sales[column] > upper_bound)].index
        outlier_values = car_sales.loc[outlier_indices, column]
        outliers[column] = outlier_values
        

for column, outlier_df in outliers.items():
    print(f"Outliers in {column}:\n{outlier_df}\n")
```

7. Make Histograms for all Numerical columns and identify the distribution of type by calculating skewnes.

```python
import matplotlib.pyplot as plt

columns = ['Sales_in_thousands', 'Price_in_thousands', 'Engine_size', 'Horsepower']

for column in columns:
    if column in car_sales.columns:
        
        plt.figure(figsize=(7, 5))
        car_sales[column].hist(bins=8)
        plt.title(f'{column}')
        plt.ylabel('Frequency')

        plt.show()

        skewness = car_sales[column].skew()
        print(f'Skewness of {column}: {skewness}\n')
```

All Histograms are skewed to the right (long right tail) with Sales_in_thousands most skewed.
